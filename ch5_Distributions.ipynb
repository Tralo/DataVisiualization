```python
iris_setosa_path='C:\\Users\\Administrator\\Desktop\\PycharmProjects\\DataVisiualization\\iris_setosa.csv'
iris_versicolor_path='C:\\Users\\Administrator\\Desktop\\PycharmProjects\\DataVisiualization\\iris_versicolor.csv'
iris_virginica_path='C:\\Users\\Administrator\\Desktop\\PycharmProjects\\DataVisiualization\\iris_virginica.csv'

iris_setosa_data = pd.read_csv(iris_setosa_path,index_col='Id')
iris_versicolor_data = pd.read_csv(iris_versicolor_path,index_col='Id')
iris_virginica_data = pd.read_csv(iris_virginica_path,index_col='Id')
```

```python
sns.distplot(a=iris_setosa_data['Petal Length (cm)'],kde=False)
sns.distplot(a=iris_versicolor_data['Petal Length (cm)'],kde=False)
sns.distplot(a=iris_virginica_data['Petal Length (cm)'],kde=False)
```

```python
sns.distplot(a=iris_setosa_data['Petal Length (cm)'],label='iris_setosa',kde=False)
sns.distplot(a=iris_versicolor_data['Petal Length (cm)'],label='iris_versicolor',kde=False)
sns.distplot(a=iris_virginica_data['Petal Length (cm)'],label='iris_virginica',kde=False)
# Add title
plt.title("Histogram of Petal Lengths, by Species")
plt.legend()
```

```python
sns.kdeplot(data=iris_setosa_data['Petal Length (cm)'],shade=True)
sns.kdeplot(data=iris_versicolor_data['Petal Length (cm)'],shade=True)
sns.kdeplot(data=iris_virginica_data['Petal Length (cm)'],shade=True)
plt.title("Histogram of Petal Lengths, by Species")
```
